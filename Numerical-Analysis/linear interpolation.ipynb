```python
from numpy import array, zeros
import matplotlib.pyplot as plt


def linearintp(x0,x1,f0,f1):
    a1=(f1-f0)/(x1-x0)
    a0=f0;
    return a0,a1
#dataset
x=array([0,1,2])

y=array([1,3,2])
n=len(x)
for i in range(n-1):
    x0=x[i]
    x1=x[i+1]
    f0=y[i]
    f1=y[i+1]
    a0,a1=linearintp(x0,x1,f0,f1)
    print('f(x) = ' ,a0,' +',a1,'(x - ' ,x0,')')
x2=0.5
y2=2
plt.scatter(x2,y2,s=60,c='w',edgecolors="k",label="x=0.5")
plt.plot([x[0],x[1]],[y[0],y[1]],'ko--',label='f0(x)')
plt.plot([x[1],x[2]],[y[1],y[2]],'ko-.',label='f1(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
